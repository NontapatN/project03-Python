```python
import pandas as pd

# Read in the data
schools = pd.read_csv("schools.csv")
schools.head()

# Create new dataframe within top 20% average math score
# Sort by average math score in descending order
best_math_school = schools[schools['average_math'] >= 0.8*800][['school_name', 'average_math']].sort_values("average_math", ascending = False)
best_math_school
```

```python
# top 10 performimg schools based on SAT
s = schools
s["total_SAT"] = s["average_math"] + s["average_reading"] + s["average_writing"]
top_10_schools = s.groupby("school_name")["total_SAT"].mean().reset_index().sort_values("total_SAT", ascending = False)
top_10_schools.head(10)
```

```python
# Find Borough with Largest Standard Deviation SAT
b = s.groupby("borough")["total_SAT"].agg(["count", "mean", "std"]).round(3)
b.head()

max_std_dev = b[b["std"] == b["std"].max()]
max_std_dev = max_std_dev.rename(columns = {"count": "number_of_schools", "mean": "average_SAT", "std": "std_SAT"})
max_std_dev
```
