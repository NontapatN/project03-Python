```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Read csv file
netflix_df = pd.read_csv("netflix_data.csv")
netflix_df.head()

# Filtering TV Show out in type column
netflix_subset = netflix_df[netflix_df["type"] != "TV Show"]
netflix_subset.head()

# Filtering only title, country, genre, release_year and duration
netflix_movies = netflix_df[["title", "country", "genre", "release_year", "duration"]]
netflix_movies.head()

# Filtering only movies with duration less than 60 minutes
short_movies = netflix_movies[netflix_movies.duration < 60]

# Append the colors for each genre for the plot
colors = []
for lab, row in netflix_movies.iterrows():
    if row["genre"] == "Children":
        colors.append("Green")
    elif row["genre"] == "Documentaries":
        colors.append("Blue")
    elif row["genre"] == "Stand-Up":
        colors.append("Red")
    else:
        colors.append("Yellow")
        
colors[:10]

# Initialize a figure and show scatter plot
fig = plt.figure(figsize = (12, 8))
plt.scatter(netflix_movies.release_year, netflix_movies.duration, c = colors)

plt.title("Movie Duration by Year of Release")
plt.xlabel("Release year")
plt.ylabel("Duration (min)")

# Movies in Netflix are getting shorter, aren't they?
answer = "Yes, movies are getting shorter"
```
